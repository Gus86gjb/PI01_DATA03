**Resoluciion**

**- Año con más carreras**

```python
carreras = pd.DataFrame(df_races['year'].value_counts())
carreras.reset_index(inplace=True)
carreras.rename({'year': 'repeticiones'}, axis=1, inplace=True)
carreras.rename({'index': 'year'}, axis=1, inplace=True)
mayor_repeticion = carreras['repeticiones'] == carreras['repeticiones'].max()
int(carreras[mayor_repeticion]['year'].to_string(index = False))
print('El año con mas carreras realizadas es', int(carreras[mayor_repeticion]['year'].to_string(index = False)))
```
